```python
import serial
```

```python
# ls /dev/tty.*
ser = serial.Serial('/dev/tty.usbserial-143120', baudrate=230400)
while True:
    data = ser.readline()[:-2] #the last bit gets rid of the new-line chars
    if data:
        print(int(data))
```

```python
from IPython.display import clear_output, display
import pandas as pd


interface_1 = '/dev/tty.usbserial-1431340'
interface_2 = '/dev/tty.usbserial-1431340'
baudrate_1 = 230400
baudrate_2 = 230400

ser = serial.Serial(interface_1, baudrate=baudrate_1)
my_list = []
df = pd.DataFrame()

try:
    while True:
        data = ser.readline()[:-2] #the last bit gets rid of the new-line chars
        
        if data:
            clear_output(wait=True)
            print(int(data))     # use display(f) if you encounter performance issues
            
            my_list.append(data)
            
except KeyboardInterrupt:
    for line in my_list:
        values = line.split('\t')
        df.append(values)
    df
    
    print('interrupted!')
    
```
