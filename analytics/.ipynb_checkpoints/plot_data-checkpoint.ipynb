```python
import numpy as np
import pandas as pd
```

```python
df_r = pd.DataFrame()
df_t = pd.DataFrame()

file = open('data/all_1567796236.txt', 'r') 

for line in file:
    values = line.split(', ')
    if values[0] == 'R':
        df_r = df_r.append({'timestamp': values[1],'r_speed': values[2],'r_pos': values[5].rstrip()}, ignore_index=True)    
    if values[0] == 'T':
        df_t = df_t.append({'timestamp': values[1],'t_pos': values[2].rstrip()}, ignore_index=True)  

file.close()

df_r['timestamp'] = pd.to_datetime(df_r['timestamp'],unit='s')
df_t['timestamp'] = pd.to_datetime(df_t['timestamp'],unit='s')

df_r = df_r.set_index('timestamp')
df_t = df_t.set_index('timestamp')

df_r['r_speed'] = df_r['r_speed'].astype(float)
df_r['r_pos'] = df_r['r_pos'].astype(float)
df_t['t_pos'] = df_t['t_pos'].astype(float)

df_r
```

```python
# synthetic data

# t_speed
t_speed = df_t.loc[df_t['t_pos'] == df_t['t_pos'], ['t_pos']]
df_t['t_speed'] = t_speed - t_speed.shift(periods=1)

# t_speed_ut1
t_speed_last = df_t['t_speed'].shift(periods=1)
filter = (df_t['t_speed'] >= -15) & (t_speed_last < -15) & (df_t['t_pos'] <= 5000)    
df_t.loc[filter, 't_speed_ut1'] = 1

# t_speed_ut2
t_speed_last = df_t['t_speed'].shift(periods=1)
filter = (df_t['t_speed'] >= 5) & (t_speed_last < 5) & (df_t['t_pos'] <= 5000)
df_t.loc[filter, 't_speed_ut2'] = 1

df_t
```

```python
# combine both df

df = pd.concat([df_r, df_t])
df = df.sort_index()

df.loc[:,['r_pos', 'r_speed', 't_pos', 't_speed']] = df.loc[:,['r_pos', 'r_speed', 't_pos', 't_speed']].fillna(method='ffill')
df
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.update_layout(
    autosize=False,
    width=1400,
    height=800
)

dfp = df[:4000]

# Add traces
fig.add_trace(go.Scatter(x=dfp.index, y=dfp['r_pos'],
                    mode='markers',
                    name='r_pos'),
            secondary_y=False)
fig.add_trace(go.Scatter(x=dfp.index, y=dfp['r_speed'],
                    mode='markers',
                    name='r_speed'),
            secondary_y=False)
fig.add_trace(go.Scatter(x=df_tp.index, y=df_tp['t_pos'],
                    mode='markers',
                    name='t_pos'),
            secondary_y=False)
fig.add_trace(go.Scatter(x=dfp.index, y=dfp['t_speed'],
                    mode='markers',
                    name='t_speed'),
            secondary_y=True)
fig.add_trace(go.Scatter(x=dfp.index, y=dfp['t_speed_ut1'],
                    mode='markers',
                    name='t_speed_ut1'),
            secondary_y=False)
fig.add_trace(go.Scatter(x=dfp.index, y=dfp['t_speed_ut2'],
                    mode='markers',
                    name='t_speed_ut2'),
            secondary_y=False)


fig.show()
```
